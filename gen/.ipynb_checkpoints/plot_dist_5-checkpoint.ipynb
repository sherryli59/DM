```python
def calc_pair_vec(particle_pos):
    pair_vec = (particle_pos.unsqueeze(-2) - particle_pos.unsqueeze(-3))
    return pair_vec

def pair_pot_hist(particle_pos, bins=100):
    dim = particle_pos.shape[-1]
    pair_vec = calc_pair_vec(particle_pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    bins = np.linspace(-0.2,0.05,50)
    pow_6 = torch.pow(sigma/distances, 6)
    distances = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)
    pot,bins = np.histogram(distances.flatten().detach().cpu().numpy(),bins=bins,density=True)
    return bins, pot

def g_r(particle_pos, bins=100):
    dim = particle_pos.shape[-1]
    pair_vec = calc_pair_vec(particle_pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    
    com = torch.mean(particle_pos,axis=1)
    dist_from_com = torch.linalg.norm(particle_pos-com.unsqueeze(1),axis=-1)
    com_atom = torch.min(dist_from_com,axis=1)[1]
    distances = distances[torch.arange(distances.shape[0]),com_atom].flatten()
    counts,bins = np.histogram(distances.detach().cpu().numpy(),bins=bins)
    bulk_density = nparticles/(4/3*np.pi*(boxlength/2)**3)
    areas = 4*np.pi*((bins[:-1]+bins[1:])/2)**2*(bins[1:]-bins[:-1])
    g_r = counts/nsamples/areas/bulk_density
    return bins, g_r
```

```python
def check_homogeneity(pos,bins=50,energy=False):
    com = torch.mean(pos,axis=1)
    dist_from_com = torch.linalg.norm(pos-com.unsqueeze(1),axis=-1)
    _,sorted_ind = torch.sort(dist_from_com,dim=-1)
    pos = torch.stack([pos[i,sorted_ind[i]] for i in range(len(pos))])
    pair_vec = calc_pair_vec(pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    
    # com_atom = torch.min(dist_from_com,axis=1)[1]
    # dist_from_com = distances[torch.arange(distances.shape[0]),com_atom].flatten()
    # sorted_dist, sorted_index = torch.sort(dist_from_com)
    #distances ,sorted_ind = torch.sort(distances,dim=-1)
    data = distances
    if energy:
        pow_6 = torch.pow(sigma/data, 6)
        data = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)
    for i in np.arange(nparticles):
        dist,bins = np.histogram(data[:,i].detach().cpu().numpy(),bins=bins,density=True)
        plt.plot(bins[:-1],dist,label="config %d"%i)
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    #pow_6 = torch.pow(sigma/torch.tensor(bins), 6)
    #pot = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)
    #plt.plot(bins[10:],pot[10:],label="potential")
    #plt.show()
```

```python
bins = np.linspace(-1,-0.1,50)
check_homogeneity(sample,bins=bins,energy=True)
check_homogeneity(xyz,bins=bins,energy=True)
plt.show()
plt.close()
bins = np.linspace(-0.1,0.2,50)
check_homogeneity(sample,bins=bins,energy=True)
check_homogeneity(xyz,bins=bins,energy=True)
plt.show()
plt.close()
bins = np.linspace(0,5,50)
check_homogeneity(sample,bins=bins)
check_homogeneity(xyz,bins=bins)
```
