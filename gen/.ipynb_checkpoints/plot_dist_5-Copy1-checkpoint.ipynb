```python
def calc_pair_vec(particle_pos):
    pair_vec = (particle_pos.unsqueeze(-2) - particle_pos.unsqueeze(-3))
    return pair_vec

def pair_pot_hist(particle_pos, bins=100):
    dim = particle_pos.shape[-1]
    pair_vec = calc_pair_vec(particle_pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    bins = np.linspace(-0.2,0.05,50)
    pow_6 = torch.pow(sigma/distances, 6)
    distances = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)
    pot,bins = np.histogram(distances.flatten().detach().cpu().numpy(),bins=bins,density=True)
    return bins, pot

def g_r(particle_pos, bins=100):
    dim = particle_pos.shape[-1]
    pair_vec = calc_pair_vec(particle_pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    
    com = torch.mean(particle_pos,axis=1)
    dist_from_com = torch.linalg.norm(particle_pos-com.unsqueeze(1),axis=-1)
    com_atom = torch.min(dist_from_com,axis=1)[1]
    distances = distances[torch.arange(distances.shape[0]),com_atom].flatten()
    counts,bins = np.histogram(distances.detach().cpu().numpy(),bins=bins)
    bulk_density = nparticles/(4/3*np.pi*(boxlength/2)**3)
    areas = 4*np.pi*((bins[:-1]+bins[1:])/2)**2*(bins[1:]-bins[:-1])
    g_r = counts/nsamples/areas/bulk_density
    return bins, g_r

def check_homogeneity(pos,bins=50,energy=False,title=None):
    com = torch.mean(pos,axis=1)
    dist_from_com = torch.linalg.norm(pos-com.unsqueeze(1),axis=-1)
    _,sorted_ind = torch.sort(dist_from_com,dim=-1)
    pos = torch.stack([pos[i,sorted_ind[i]] for i in range(len(pos))])
    pair_vec = calc_pair_vec(pos)
    nsamples = len(pair_vec)
    nparticles = pair_vec.shape[1]
    distances = torch.linalg.norm(pair_vec.float(), axis=-1)
    # remove diagonal zeros
    rem_dims = distances.shape[:-2]
    distances = distances.flatten(start_dim=-2)[...,1:].view(
    *rem_dims,nparticles-1, nparticles+1)[...,:-1].reshape(*rem_dims,nparticles, nparticles-1)
    data = distances
    if energy:
        pow_6 = torch.pow(sigma/data, 6)
        data = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)
    for i in np.arange(nparticles):
        dist,bins = np.histogram(data[:,i].detach().cpu().numpy(),bins=bins,density=True)
        plt.plot(bins[:-1],dist,label="config %d"%i)
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    if title is not None:
        plt.title(title)
    plt.show()
```

```python
def analyze(cfg):
    sigma = 1
    epsilon = 1
    boxlength = 5.42
    nparticles = 50
    sys = lj.LJ(boxlength=0,cutoff=None,shift=False,periodic=False)
    #data_dir = "/home/sherryli/xsli/GenerativeModel/data/lj/liquid_3d/nonperiodic/lj_5_3d_2.0box_T1.00/"
    data_dir = "/home/sherryli/xsli/GenerativeModel/data/lj/liquid_3d/nonperiodic/lj_50_3d_3.0box_T1.00/"

    traj = md.load_hdf5(data_dir+"run0/lj.h5")
    xyz = torch.tensor(np.load(data_dir+"pos_test.npy"),dtype=torch.float64)
    sim_pot = sys.potential(xyz)
    model_dict=load_model(cfg,name=cfg.dataset.name,model_dir=cfg.output.training_dir,mode="sample")
    model = model_dict["ema_model"]
    score = model.score_fn
    nsamples = 1000
    output = model.sample(nsamples=nsamples,init=xyz[:nsamples].to("cuda:0"), 
        return_prob=False,method="ode",return_traj=False)
    sample = output["x"]
    generated_pot = sys.potential(sample)
    print(torch.mean(generated_pot))
    generated_pot = generated_pot[generated_pot <100]
    plt.hist(sim_pot.detach().cpu().numpy(),bins=50,range=(-4,1),density=True,alpha=0.5,label="sim")
    plt.hist(generated_pot.detach().cpu().numpy(),range=(-3,1),bins=50,density=True,alpha=0.5,label="generated")
    plt.legend()
    plt.show()
    
    
    bins = np.linspace(-1,5,50)
    check_homogeneity(sample,bins=bins,title="sample,distance")
    check_homogeneity(xyz,bins=bins,title="sim,distance")
    bins = np.linspace(-0.1,0.2,50)
    check_homogeneity(sample,bins=bins,energy=True,title="sample,energy")
    check_homogeneity(xyz,bins=bins,energy=True,title="sim,energy")
    
    
    bins,gr_sim = g_r(xyz)
    bins_generated,gr_generated = g_r(sample)

    pow_6 = torch.pow(sigma/torch.tensor(bins), 6)
    pot = 4*epsilon * (torch.pow(pow_6, 2)- pow_6)

    plt.plot(bins_generated[:-1],gr_generated,label="generated")
    plt.plot(bins[:-1],gr_sim,label="sim")
    plt.plot(bins,pot,label="LJ potential for ref")
    plt.title("g_r")
    plt.legend()
    plt.show()
    return model,sample,xyz
```
